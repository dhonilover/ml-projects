```python
#removing skewness using cuberoot method

df["Malic acid"]=np.cbrt(df["Malic acid"])
df["Alcalinity of ash"]=np.cbrt(df["Alcalinity of ash"])
df["Magnesium"]=np.cbrt(df["Magnesium"])
df["Flavanoids"]=np.cbrt(df["Flavanoids"])
df["Nonflavanoid phenols"]=np.cbrt(df["Nonflavanoid phenols"])
df["diluted wines"]=np.cbrt(df["diluted wines"])
```

# Lets Check Variance Inflation Factor

```python
#finding variance inflation factor in each scaled column 
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["VIF values"] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
vif["Features"] = x.columns
vif
```

Note- we can see that  vif values for all columns is in permissible limit

Now we have the data  balanced, so lets make a machine learning model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso,Ridge

maxAccu = 0
maxRS = 0
```

```python
# finding the best accuracy state


for i in range(1,96):
    x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.30, random_state = i)
    lr=LinearRegression()
    lr.fit(x_train, y_train)
    pred= lr.predict(x_test)
    acc= r2_score(y_test,pred)
    
    if acc>maxAccu:
        maxAccu=acc
        maxRS = i
        
print("max r2_score  is ", maxAccu, "at random_state", maxRS)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param={'alpha':[1.0,.05,.4,2],'fit_intercept':[True,False],'solver':['auto','svd','cholesky','lsqr','sag','saga','lbfgs'],'positive':[False,True],'random_state':[1,4,10,20]
      
      }
gscv=GridSearchCV(Ridge(),param,cv=5)
gscv.fit(x_train,y_train)
```

```python
gscv.best_params_
```

```python
Model=Ridge(alpha=.05,fit_intercept=True,positive=False,random_state=1,solver='lsqr')
```

```python
Model.fit(x_train,y_train)
pred=Model.predict(x_test)
print('r2_score:',r2_score(y_test,pred))
```
